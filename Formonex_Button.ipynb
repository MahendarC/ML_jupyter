```python
import requests
from bs4 import BeautifulSoup
```

```python
baseurl = 'https://mrbutton.in/collections/shirts'
headers = {'User-Agent': 'Mozilla/4.0 (compatible; MSIE 6.0; AOL 9.0; Windows NT 5.1; {}; Alawar 2.08; .NET CLR 1.0.3705)'}
```

```python
req = requests.get(baseurl, headers=headers)
soup = BeautifulSoup(req.content, 'html')
```

```python
products = soup.find_all('div', class_='grid__item')
```

```python
links = []

for item in products:
    for link in item.find_all('a', href=True):
        links.append('https://mrbutton.in/' + link['href'])
```

```python
links
```

```python
shirtlist=[]
for link in links:
    try:
        r = requests.get(link, headers=headers)
        soup = BeautifulSoup(r.content, 'html')
        name = soup.find('h1', class_='h2 product-single__title').text.strip()
        MRP = soup.find('span', class_='product__price--compare').text.strip()
        SP = soup.find('span', class_='product__price on-sale').text.strip()
        
        print(name)
        shirts = {
            'name':name,
            'MRP':MRP,
            'SP':SP
        }
        shirtlist.append(shirts)
    except:
        continue
```

```python
import pandas as pd
df = pd.DataFrame(shirtlist)
```

```python
df.to_csv('MrButton.csv')
```
